```python
df = pd.read_csv(
    "../data/consolas_vendidas_limpio.csv", 
    sep = ";",
    decimal = ".",
    encoding="latin-1")
display(df)
```

Obtención de los datos de media, mediana y moda del total de las consolas según región, Después obtendremos el global.

```python
media_ven_na = df['North America'].mean()
print("La media de consolas vendidas en Norteamérica es:", round(media_ven_na, 2))
mediana_ven_na = df['North America'].median()
print("La mediana de consolas vendidas en Norteamérica es:", mediana_ven_na)
moda_ven_na = df['North America'].mode()[0]
print("El valor mas recurrente en Norteamérica es:", moda_ven_na)

media_ven_eu = df['Europe'].mean()
print("La media de consolas vendidas en Europa es:", round(media_ven_eu, 2))
mediana_ven_eu = df['Europe'].median()
print("La mediana de consolas vendidas en Europa es:", mediana_ven_eu)
moda_ven_eu = df['Europe'].mode()[0]
print("El valor mas recurrente en Europa es:", moda_ven_eu)

media_ven_jp = df['Japan'].mean()
print("La media de consolas vendidas en Japón es:", round(media_ven_jp, 2))
mediana_ven_jp = df['Japan'].median()
print("La mediana de consolas vendidas en Japón es:", mediana_ven_jp)
moda_ven_ja = df['Japan'].mode()[0]
print("El valor mas recurrente en Japón es:", moda_ven_ja)

media_ven_ot = df['Rest of World'].mean()
print("La media de consolas vendidas en el resto del mundo es:", round(media_ven_ot, 2))
mediana_ven_ot = df['Rest of World'].median()
print("La mediana de consolas vendidas en el resto del mundo es:", mediana_ven_ot)
moda_ven_ot = df['Rest of World'].mode()[0]
print("El valor mas recurrente en el resto del mundo es:", moda_ven_ot)
```

```python
media_ven_gl = df['Global'].mean()
print("La media de consolas vendidas en general es:", round(media_ven_gl, 2))
mediana_ven_gl = df['Global'].median()
print("La mediana de consolas vendidas en general es:", mediana_ven_gl)
moda_ven_gl = df['Global'].mode()[0]
print("El valor mas recurrente en general es:", moda_ven_gl)
```

Hacemos las graficas de los datos obtenidos

```python
datos_grafica = {
    'Región': ['Norteamérica', 'Europa', 'Japón', 'Resto del Mundo'],
    'Media Ventas': [media_ven_na, media_ven_eu, media_ven_jp, media_ven_ot],
    "Mediana Ventas": [mediana_ven_na, mediana_ven_eu, mediana_ven_jp, mediana_ven_ot],
    "Moda Ventas": [moda_ven_na, moda_ven_eu, moda_ven_ja, moda_ven_ot]
}
df_stats = pd.DataFrame(datos_grafica)

plt.figure(figsize=(10,6))
plt.bar(df_stats['Región'], df_stats['Media Ventas'], color=['blue', 'orange', 'green', 'red']) 
plt.xlabel('Región')
plt.ylabel('Media de Consolas Vendidas')
plt.title('Media de Ventas de Consolas por Región')
plt.show() 
```

```python
plt.figure(figsize=(10,6))
plt.bar(df_stats['Región'], df_stats['Mediana Ventas'], color=['blue', 'orange', 'green', 'red']) 
plt.xlabel('Región')
plt.ylabel('Mediana de Consolas Vendidas')
plt.title('Mediana de Ventas de Consolas por Región')
plt.show() 
```

```python
plt.figure(figsize=(10,6))
plt.bar(df_stats['Región'], df_stats['Moda Ventas'], color=['blue', 'orange', 'green', 'red']) 
plt.xlabel('Región')
plt.ylabel('Moda de Consolas Vendidas')
plt.title('Moda de Ventas de Consolas por Región')
plt.show() 
```

```python

datos_global = {
    'Región': ['Global', 'Global', 'Global'],
    'Ventas': [media_ven_gl,mediana_ven_gl ,moda_ven_gl]    
}
df_stats_global = pd.DataFrame(datos_global, index=['Media', 'Mediana', 'Moda'])
plt.figure(figsize=(10,6))
plt.bar(df_stats_global.index, df_stats_global['Ventas'], color=['blue', 'orange', 'green']) 
plt.xlabel('Ventas')
plt.ylabel('Global Consolas Vendidas')
plt.title('Global de Ventas de Consolas')
plt.show() 
```

3. Analisis multivariable

```python
cols = ['Global', 'Europe', 'North America', 'Japan']
```

```python
sns.pairplot(df[cols])
plt.title('Gráfico Ventas Multivariante')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df[cols].corr() 
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
plt.title('Mapa de calor de la correlacion de las ventas')
plt.show()
```
